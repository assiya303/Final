```python
import datetime as dt
import pandas as pd
```

```python
nba_stats = pd.read_csv('Raw_Compiled_Stats.csv')
nba_stats.head()
```

```python
nba_stats.columns
```

```python
nba_stats.rename(columns = {list(nba_stats)[2]: 'Game Date'}, inplace = True)
```

```python
nba_stats['Game Date'] = pd.to_datetime(nba_stats['Game Date'])
nba_stats.head()
```

```python
nba = nba_stats[['Team', 'Game Date', 'PTS', 'FG%', '3P%', 'FT%', 'OREB', 'DREB', 'REB', 'AST', 'BLK', 'TOV', 'PF']]
nba.head()
```

```python
start_1314 = '10-28-2013'
end_1314 = '04-17-2014'

start_1415 = '10-27-2014'
end_1415 = '04-16-2015'

start_1516 = '10-26-2015'
end_1516 = '04-14-2016'

start_1617 = '10-24-2016'
end_1617 = '04-13-2017'

start_1718 = '10-16-2017'
end_1718 = '04-12-2018'
```

```python
mask_1314 = (nba['Game Date'] > start_1314) & (nba['Game Date'] <= end_1314)

mask_1415 = (nba['Game Date'] > start_1415) & (nba['Game Date'] <= end_1415)

mask_1516 = (nba['Game Date'] > start_1516) & (nba['Game Date'] <= end_1516)

mask_1617 = (nba['Game Date'] > start_1617) & (nba['Game Date'] <= end_1617)

mask_1718 = (nba['Game Date'] > start_1718) & (nba['Game Date'] <= end_1718)
```

```python
avg_1314 = nba.loc[mask_1314]
avg_1314.head()
```

```python
avg_1314 = avg_1314.groupby(['Team']).mean()
avg_1314.head()
```

```python
avg_1314.reset_index(inplace=True)
avg_1314.head()
```

```python
avg_1314.columns
```

```python
avg_1314.rename(columns = {list(avg_1314)[0]: 'TEAM'}, inplace = True)
avg_1314.head()
```

```python
home_team = avg_1314['TEAM']
print(home_team.head())

away_team = avg_1314['TEAM'].shift(periods =-1)
print(away_team)
```

```python
stats_1314 = avg_1314.drop(labels='TEAM', axis=1)
stats_1314.head()
```

```python
home = []

for z in range(1,30):
    home_team = avg_1314.iloc[:,0]
    home_team.drop(home_team.tail(z).index, inplace=True)
    home.append(home_team)

team_home = pd.concat(home, axis=0, sort=True)
team_home.head()
```

```python
away = []

for y in range(1,30):
    away_team = avg_1314.iloc[:,0].shift(periods =- y)
    away_team.drop(away_team.tail(y).index, inplace=True)
    away.append(away_team)

team_away = pd.concat(away, axis=0, sort=True)
team_away.head()
```

```python
diff = []

for x in range(1, 30):
    diff_1314 = stats_1314 - stats_1314.shift(periods =- x)
    diff_1314.drop(diff_1314.tail(x).index, inplace=True)
    home_team = avg_1314.iloc[:,0]
    home_team.drop(home_team.tail(x).index, inplace=True)
    away_team = avg_1314.iloc[:,0].shift(periods =- x)
    away_team.drop(away_team.tail(x).index, inplace=True)
    diff_1314['AWAY'] = away_team
    diff_1314['HOME'] = home_team
    diff.append(diff_1314)

diff_reg_1314 = pd.concat(diff, axis=0, sort=True)
diff_reg_1314.head()
```

```python
matchup_1314 = diff_reg_1314[['HOME', 'AWAY', 'PTS', 'FG%', '3P%', 'FT%', 'OREB', 'DREB', 'REB', 'AST', 'BLK', 'TOV', 'PF']]
matchup_1314.head()
```

```python
# matchup_1314.drop('Test Neg Pts', axis=1, inplace=True)
neg_matchup_1314 = matchup_1314.iloc[:,2:13].multiply(-1)
matchup_1314.head()
```

```python
neg_matchup_1314['HOME'] = matchup_1314['AWAY']
neg_matchup_1314['AWAY'] = matchup_1314['HOME']
neg_matchup_1314.head()
```
