```python
from splinter import Browser
from bs4 import BeautifulSoup
import pandas as pd
```

```python
def init_browser():
    executable_path = {'executable_path': 'chromedriver.exe'}
    return Browser('chrome', **executable_path, headless=False)
```

```python
browser = init_browser()

#list url to visit
season = 'ALLTIME'
season_type = 'Regular%20Season'
date_from = '10%2F01%2F2013'
date_to = '04%2F30%2F2018'

nba_url = 'https://stats.nba.com/teams/boxscores/?Season={season}&SeasonType={season_type}&DateFrom={date_from}&DateTo={date_to}'.format(season=season, season_type=season_type, date_from=date_from, date_to=date_to)

#visit url
browser.visit(nba_url)
print(nba_url)
```

```python
full = browser.find_by_tag('nba-stat-table')
full_t = full.find_by_tag('select')[0]
full_table = full_t.find_by_tag('option')[0]
full_table.click()
```

```python
nba_html = browser.html 
soup = BeautifulSoup(nba_html, 'html.parser')
```

```python
table = soup.find_all('nba-stat-table')[0]
```

```python
nba_stats = pd.read_html(str(table), header=0)[0]
nba_stats
```

```python
nba_stats.to_csv('Raw_Compiled_Stats.csv')
```

```python
nba_stats.columns
```

```python
nba_stats_clean = nba_stats[['PTS', 'FG%', '3P%', 'FT%', 'OREB', 'DREB', 'REB', 'AST', 'BLK', 'TOV', 'PF']]
nba_stats_clean.head(10)
```

```python
nba_new = nba_stats_clean - nba_stats_clean.shift(periods =-1)
nba_new.head(10)
```

```python
nba_diff = nba_new.iloc[::2]
nba_diff.head(10)
```

```python
nba_diff.to_csv('NBA_Differential.csv')
```

```python
nba_diff = pd.read_csv('NBA_Differential.csv')
nba_diff.drop(nba_diff.columns[0], axis=1, inplace=True)
nba_diff.head(10)
```

```python
X=nba_diff.drop('PTS', axis=1)
y=nba_diff['PTS']
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
model.fit(X, y)

training_score = model.score(X, y)
testing_score = model.score(X_test, y_test)

print(f"Training Score: {training_score}")
print(f"Testing Score: {testing_score}")
```

```python
import matplotlib.pyplot as plt

plt.scatter(model.predict(X_train), model.predict(X_train) - y_train, c="blue", label="Training Data")
plt.scatter(model.predict(X_test), model.predict(X_test) - y_test, c="orange", label="Testing Data")
plt.legend()
plt.hlines(y=0, xmin=y.min(), xmax=y.max())
plt.title("Residual Plot")
plt.show
```

```python
from sklearn.metrics import mean_squared_error, r2_score

predicted = model.predict(X)
```

```python
mse = mean_squared_error(y, predicted)
r2 = r2_score(y, predicted)

print(f'Mean Squared Error: {mse}')
print(f'R-Squared: {r2}')
print(f'Intercept: {model.intercept_}')
print(f'Coefficients: {model.coef_}')
print(X.columns)
```

Deep Learning Model

```python
X = nba_stats_clean.drop('PTS', axis=1)
y = nba_stats_clean['PTS']
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from keras.utils import to_categorical
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify=y)
X_scaler = StandardScaler().fit(X_train)
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
#Label-encode data set
label_encoder = LabelEncoder()
label_encoder.fit(y_train)
encoded_y_train = label_encoder.transform(y_train)
encoded_y_test = label_encoder.transform(y_test)
```

```python
#Convert encoded labels to one-hot-encoding
y_train_categorical = to_categorical(encoded_y_train)
y_test_categorical = to_categorical(encoded_y_test)
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
# Create a Neural Network model here
deep_model = Sequential()
number_inputs = 14
number_hidden_nodes = 100
deep_model.add(Dense(units=number_hidden_nodes, activation='relu', input_dim=number_inputs))
deep_model.add(Dense(units=number_hidden_nodes, activation='relu'))
deep_model.add(Dense(units=2, activation='softmax'))
```

```python
deep_model.summary()
```

```python
# Compile and fit the model
deep_model.compile(optimizer='adam',
                   loss='categorical_crossentropy',
                   metrics=['accuracy'])

deep_model.fit(
    X_train_scaled,
    y_train_categorical,
    epochs=100,
    shuffle=True,
    verbose=2
)
```

```python
model_loss, model_accuracy = deep_model.evaluate(X_test_scaled, y_test_categorical, verbose=2)
print(f"Normal Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Use the first 5 test data values to make a prediction and compare it to the actual labels
encoded_predictions = deep_model.predict_classes(X_test_scaled[:5])
prediction_labels = label_encoder.inverse_transform(encoded_predictions)

print(f'Predicted classes: {prediction_labels}')
print(f"Actual Labels: {list(y_test[:5])}")
```

```python
deep_model.save('nba_game_predictor.h5')
```
